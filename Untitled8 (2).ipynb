```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import fpgrowth, association_rules
from mlxtend.preprocessing import TransactionEncoder
from collections import Counter
import matplotlib.pyplot as plt
import seaborn as sns
from itertools import combinations
import time
```

```python

# Set random seed for reproducibility
np.random.seed(42)

def create_dummy_dataset(n_rows=10000, n_cols=900, sparsity=0.95):
    """
    Create a dummy one-hot encoded dataset with realistic patterns

    Args:
        n_rows: Number of rows (samples)
        n_cols: Number of columns (flags)
        sparsity: Proportion of zeros (higher = more sparse)
    """
    print(f"Creating dummy dataset: {n_rows} rows × {n_cols} columns")

    # Create column names
    flag_names = [f'flag_{i+1}' for i in range(n_cols)]

    # Generate data with some realistic patterns
    # Method 1: Pure random with sparsity
    data = np.random.binomial(1, 1-sparsity, size=(n_rows, n_cols))

    # Method 2: Add some correlated patterns to make it more realistic
    # Create some groups of flags that tend to occur together
    n_pattern_groups = 20
    pattern_size = 5
    pattern_strength = 0.3

    for _ in range(n_pattern_groups):
        # Select random flags for this pattern
        pattern_flags = np.random.choice(n_cols, size=pattern_size, replace=False)
        # Select random samples to apply this pattern
        pattern_samples = np.random.choice(n_rows, size=int(n_rows * pattern_strength), replace=False)

        # Set these flags to 1 for selected samples
        for flag_idx in pattern_flags:
            data[pattern_samples, flag_idx] = 1

    # Convert to DataFrame
    df = pd.DataFrame(data, columns=flag_names)

    print(f"Dataset created successfully!")
    print(f"Shape: {df.shape}")
    print(f"Actual sparsity: {(df == 0).sum().sum() / (df.shape[0] * df.shape[1]):.3f}")
    print(f"Total active flags per row - Mean: {df.sum(axis=1).mean():.2f}, Std: {df.sum(axis=1).std():.2f}")

    return df
```

```python


def analyze_basic_patterns(df, top_n=20):
    """
    Analyze basic flag frequencies and combinations
    """
    print("\n" + "="*60)
    print("BASIC PATTERN ANALYSIS")
    print("="*60)

    # Individual flag frequencies
    flag_counts = df.sum().sort_values(ascending=False)
    print(f"\nTop {top_n} most frequent individual flags:")
    print(flag_counts.head(top_n))

    # Flag activation statistics
    row_sums = df.sum(axis=1)
    print(f"\nFlag activation per sample statistics:")
    print(f"Mean flags per sample: {row_sums.mean():.2f}")
    print(f"Median flags per sample: {row_sums.median():.2f}")
    print(f"Max flags per sample: {row_sums.max()}")
    print(f"Min flags per sample: {row_sums.min()}")

    return flag_counts
```

```python

def efficient_frequent_patterns_fpgrowth(df, min_support=0.01, max_len=5):
    """
    Use FP-Growth algorithm for efficient frequent pattern mining
    This is more efficient than Apriori for sparse datasets
    """
    print(f"\n" + "="*60)
    print("FP-GROWTH FREQUENT PATTERN MINING")
    print("="*60)
    print(f"Parameters: min_support={min_support}, max_len={max_len}")

    start_time = time.time()

    # Convert to boolean for mlxtend
    df_bool = df.astype(bool)

    # Apply FP-Growth
    print("Running FP-Growth algorithm...")
    frequent_itemsets = fpgrowth(df_bool, min_support=min_support, max_len=max_len, use_colnames=True)

    if len(frequent_itemsets) == 0:
        print(f"No frequent patterns found with min_support={min_support}")
        print("Try reducing the min_support threshold")
        return None

    # Sort by support (frequency)
    frequent_itemsets = frequent_itemsets.sort_values('support', ascending=False)

    # Add pattern length and actual count
    frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(len)
    frequent_itemsets['count'] = (frequent_itemsets['support'] * len(df)).astype(int)

    end_time = time.time()
    print(f"FP-Growth completed in {end_time - start_time:.2f} seconds")
    print(f"Found {len(frequent_itemsets)} frequent patterns")

    return frequent_itemsets
```

```python

def analyze_pattern_results(frequent_itemsets, top_n=20):
    """
    Analyze and display the frequent pattern results
    """
    if frequent_itemsets is None or len(frequent_itemsets) == 0:
        print("No patterns to analyze")
        return

    print(f"\n" + "="*60)
    print("PATTERN ANALYSIS RESULTS")
    print("="*60)

    # Overall statistics
    print(f"Total frequent patterns found: {len(frequent_itemsets)}")
    print(f"Pattern length distribution:")
    length_dist = frequent_itemsets['length'].value_counts().sort_index()
    for length, count in length_dist.items():
        print(f"  Length {length}: {count} patterns")

    # Top patterns overall
    print(f"\nTop {top_n} Most Frequent Patterns (All Lengths):")
    print("-" * 80)
    for idx, row in frequent_itemsets.head(top_n).iterrows():
        pattern_str = ', '.join(sorted(row['itemsets']))
        print(f"Support: {row['support']:.4f} | Count: {row['count']:4d} | Length: {row['length']} | Pattern: {{{pattern_str}}}")

    # Top patterns by length
    for length in sorted(frequent_itemsets['length'].unique()):
        if length == 1:
            continue  # Skip individual flags as we analyzed them separately

        length_patterns = frequent_itemsets[frequent_itemsets['length'] == length].head(10)
        if len(length_patterns) > 0:
            print(f"\nTop 10 Patterns of Length {length}:")
            print("-" * 50)
            for idx, row in length_patterns.iterrows():
                pattern_str = ', '.join(sorted(row['itemsets']))
                print(f"  Support: {row['support']:.4f} | Count: {row['count']:4d} | {{{pattern_str}}}")
```

```python

def find_association_rules(frequent_itemsets, min_confidence=0.5):
    """
    Generate association rules from frequent patterns
    """
    if frequent_itemsets is None or len(frequent_itemsets) == 0:
        print("No frequent itemsets available for association rule mining")
        return None

    print(f"\n" + "="*60)
    print("ASSOCIATION RULES")
    print("="*60)

    try:
        # Filter for patterns with length > 1
        multi_item_patterns = frequent_itemsets[frequent_itemsets['length'] > 1]

        if len(multi_item_patterns) == 0:
            print("No multi-item patterns found for association rule mining")
            return None

        # Generate association rules
        rules = association_rules(multi_item_patterns, metric="confidence", min_threshold=min_confidence)

        if len(rules) == 0:
            print(f"No association rules found with min_confidence={min_confidence}")
            return None

        # Sort by confidence and lift
        rules = rules.sort_values(['confidence', 'lift'], ascending=False)

        print(f"Found {len(rules)} association rules")
        print(f"\nTop 15 Association Rules (sorted by confidence):")
        print("-" * 100)

        for idx, rule in rules.head(15).iterrows():
            antecedent = ', '.join(sorted(rule['antecedents']))
            consequent = ', '.join(sorted(rule['consequents']))
            print(f"{{<b>{antecedent}</b>}} → {{<b>{consequent}</b>}} | "
                  f"Conf: {rule['confidence']:.3f} | "
                  f"Lift: {rule['lift']:.3f} | "
                  f"Support: {rule['support']:.3f}")

        return rules

    except Exception as e:
        print(f"Error generating association rules: {e}")
        return None
```

```python

def create_visualizations(df, frequent_itemsets, flag_counts):
    """
    Create visualizations for the pattern analysis
    """
    plt.figure(figsize=(15, 10))

    # 1. Flag frequency distribution
    plt.subplot(2, 3, 1)
    top_flags = flag_counts.head(20)
    plt.bar(range(len(top_flags)), top_flags.values)
    plt.title('Top 20 Individual Flag Frequencies')
    plt.xlabel('Flag Rank')
    plt.ylabel('Frequency')
    plt.xticks(range(0, len(top_flags), 5))

    # 2. Flags per sample distribution
    plt.subplot(2, 3, 2)
    row_sums = df.sum(axis=1)
    plt.hist(row_sums, bins=50, alpha=0.7)
    plt.title('Distribution of Active Flags per Sample')
    plt.xlabel('Number of Active Flags')
    plt.ylabel('Number of Samples')

    # 3. Pattern length distribution
    if frequent_itemsets is not None and len(frequent_itemsets) > 0:
        plt.subplot(2, 3, 3)
        length_dist = frequent_itemsets['length'].value_counts().sort_index()
        plt.bar(length_dist.index, length_dist.values)
        plt.title('Frequent Pattern Length Distribution')
        plt.xlabel('Pattern Length')
        plt.ylabel('Number of Patterns')

        # 4. Support distribution
        plt.subplot(2, 3, 4)
        plt.hist(frequent_itemsets['support'], bins=30, alpha=0.7)
        plt.title('Support Distribution of Frequent Patterns')
        plt.xlabel('Support')
        plt.ylabel('Number of Patterns')

        # 5. Top patterns by support
        plt.subplot(2, 3, 5)
        top_patterns = frequent_itemsets.head(15)
        plt.barh(range(len(top_patterns)), top_patterns['support'])
        plt.title('Top 15 Patterns by Support')
        plt.xlabel('Support')
        plt.ylabel('Pattern Rank')

    plt.tight_layout()
    plt.show()
```

```python

def memory_efficient_analysis(df, min_support_start=0.01, min_support_step=0.005):
    """
    Memory-efficient analysis that gradually reduces min_support if no patterns are found
    """
    min_support = min_support_start
    frequent_itemsets = None

    while frequent_itemsets is None or len(frequent_itemsets) == 0:
        print(f"\nTrying min_support = {min_support:.4f}")
        frequent_itemsets = efficient_frequent_patterns_fpgrowth(df, min_support=min_support, max_len=4)

        if frequent_itemsets is not None and len(frequent_itemsets) > 0:
            break

        min_support -= min_support_step
        if min_support <= 0.001:
            print("Reached minimum support threshold. No frequent patterns found.")
            break

    return frequent_itemsets
```

```python

# ================================
# MAIN EXECUTION
# ================================

def main():
    print("FREQUENT PATTERN MINING FOR ONE-HOT ENCODED DATA")
    print("=" * 60)

    # Create dummy dataset
    print("Step 1: Creating dummy dataset...")
    X_train = create_dummy_dataset(n_rows=1000, n_cols=200, sparsity=0.95)
    # X_train=X_train[:10]
    # Basic analysis
    print("\nStep 2: Basic pattern analysis...")
    flag_counts = analyze_basic_patterns(X_train, top_n=20)

    # Frequent pattern mining with adaptive min_support
    print("\nStep 3: Frequent pattern mining...")
    frequent_itemsets = memory_efficient_analysis(X_train, min_support_start=0.005)

    # Analyze results
    if frequent_itemsets is not None and len(frequent_itemsets) > 0:
        print("\nStep 4: Analyzing pattern results...")
        analyze_pattern_results(frequent_itemsets, top_n=25)

        print("\nStep 5: Finding association rules...")
        rules = find_association_rules(frequent_itemsets, min_confidence=0.3)

        print("\nStep 6: Creating visualizations...")
        create_visualizations(X_train, frequent_itemsets, flag_counts)

        # Save results
        print("\nStep 7: Saving results...")
        frequent_itemsets.to_csv('frequent_patterns.csv', index=False)
        if rules is not None:
            rules.to_csv('association_rules.csv', index=False)
        print("Results saved to CSV files")

        return X_train, frequent_itemsets, rules
    else:
        print("No frequent patterns found. Consider:")
        print("1. Reducing the min_support threshold further")
        print("2. Checking if your data is too sparse")
        print("3. Increasing the pattern occurrence in your dataset")

        return X_train, None, None

# Run the analysis
if __name__ == "__main__":
    # Install required packages first
    print("Required packages: pandas, numpy, mlxtend, matplotlib, seaborn")
    print("Install with: pip install pandas numpy mlxtend matplotlib seaborn")
    print("\n" + "="*60)

    X_train, patterns, rules = main()

    print(f"\nAnalysis complete!")
    print(f"Dataset shape: {X_train.shape}")
    if patterns is not None:
        print(f"Frequent patterns found: {len(patterns)}")
    if rules is not None:
        print(f"Association rules found: {len(rules)}")
```

```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import fpgrowth
import time
import gc

def create_memory_efficient_dataset(n_rows=10000, n_cols=900, sparsity=0.95):
    """
    Create dataset with optimal memory usage for 900 columns
    """
    print(f"Creating dataset: {n_rows} rows × {n_cols} columns")

    # Use most memory-efficient data type
    data = np.random.binomial(1, 1-sparsity, size=(n_rows, n_cols)).astype(np.bool_)

    # Add realistic patterns without reducing columns
    n_pattern_groups = 100  # More patterns across all 900 columns

    for i in range(n_pattern_groups):
        pattern_size = np.random.randint(2, 6)
        # Ensure patterns span across all 900 columns
        pattern_flags = np.random.choice(n_cols, size=pattern_size, replace=False)
        pattern_samples = np.random.choice(n_rows, size=int(n_rows * np.random.uniform(0.05, 0.15)), replace=False)
        data[np.ix_(pattern_samples, pattern_flags)] = 1

    flag_names = [f'flag_{i+1}' for i in range(n_cols)]
    df = pd.DataFrame(data, columns=flag_names, dtype=bool)

    print(f"Memory usage: {df.memory_usage(deep=True).sum() / 1024**2:.1f} MB")
    return df

def optimized_fpgrowth_full_columns(df, min_support=0.005, max_len=4):
    """
    FP-Growth optimized for full 900-column dataset
    """
    print(f"Running FP-Growth on full dataset: {df.shape}")
    print(f"Parameters: min_support={min_support}, max_len={max_len}")

    start_time = time.time()

    # Key optimization: Use boolean dtype directly (already done in dataset creation)
    print("Dataset already in optimal boolean format")

    # Calculate minimum count threshold
    min_count = int(min_support * len(df))
    print(f"Minimum count threshold: {min_count} transactions")

    try:
        # Run FP-Growth with memory-conscious settings
        frequent_itemsets = fpgrowth(
            df,
            min_support=min_support,
            max_len=max_len,
            use_colnames=True
        )

        if len(frequent_itemsets) == 0:
            print(f"No patterns found with min_support={min_support}")
            print("Try reducing min_support or check data sparsity")
            return None

        # Add analysis columns efficiently
        frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(len)
        frequent_itemsets['count'] = (frequent_itemsets['support'] * len(df)).astype(int)
        frequent_itemsets = frequent_itemsets.sort_values('support', ascending=False)

        end_time = time.time()
        print(f"FP-Growth completed in {end_time - start_time:.2f} seconds")
        print(f"Found {len(frequent_itemsets)} frequent patterns across all 900 columns")

        return frequent_itemsets

    except MemoryError as e:
        print(f"Memory error: {e}")
        print("Suggestions:")
        print("1. Increase min_support (try 0.01 or 0.02)")
        print("2. Reduce max_len to 3")
        print("3. Process in batches")
        return None

def adaptive_parameter_tuning(df):
    """
    Smart parameter selection for 900-column dataset
    """
    n_rows, n_cols = df.shape
    sparsity = (df == False).sum().sum() / (n_rows * n_cols)
    avg_items_per_row = df.sum(axis=1).mean()
    max_items_per_row = df.sum(axis=1).max()

    print(f"Dataset analysis for parameter tuning:")
    print(f"  Rows: {n_rows}, Columns: {n_cols}")
    print(f"  Sparsity: {sparsity:.3f}")
    print(f"  Avg active flags per row: {avg_items_per_row:.2f}")
    print(f"  Max active flags per row: {max_items_per_row}")

    # Conservative parameters for 900 columns to avoid timeout
    if avg_items_per_row > 50:  # Dense transactions
        min_support = max(0.02, 50 / n_rows)
        max_len = 3
    elif avg_items_per_row > 20:  # Medium density
        min_support = max(0.01, 25 / n_rows)
        max_len = 4
    else:  # Sparse transactions
        min_support = max(0.005, 10 / n_rows)
        max_len = 5

    print(f"Recommended parameters: min_support={min_support:.4f}, max_len={max_len}")
    return min_support, max_len

def comprehensive_pattern_analysis(frequent_itemsets, df):
    """
    Analyze patterns across all 900 columns
    """
    if frequent_itemsets is None or len(frequent_itemsets) == 0:
        print("No patterns to analyze")
        return

    print(f"\n" + "="*60)
    print("COMPREHENSIVE PATTERN ANALYSIS (ALL 900 COLUMNS)")
    print("="*60)

    # Column coverage analysis
    all_items_in_patterns = set()
    for itemset in frequent_itemsets['itemsets']:
        all_items_in_patterns.update(itemset)

    coverage = len(all_items_in_patterns) / len(df.columns)
    print(f"Column coverage: {len(all_items_in_patterns)}/{len(df.columns)} ({coverage:.1%})")

    # Pattern length distribution
    length_dist = frequent_itemsets['length'].value_counts().sort_index()
    print(f"\nPattern length distribution:")
    for length, count in length_dist.items():
        avg_support = frequent_itemsets[frequent_itemsets['length'] == length]['support'].mean()
        print(f"  Length {length}: {count} patterns (avg support: {avg_support:.4f})")

    # Top patterns by length
    print(f"\nTop patterns by length:")
    for length in sorted(frequent_itemsets['length'].unique()):
        length_patterns = frequent_itemsets[frequent_itemsets['length'] == length].head(5)
        print(f"\n  Top 5 patterns of length {length}:")
        for idx, row in length_patterns.iterrows():
            pattern_str = ', '.join(sorted(row['itemsets']))
            print(f"    Support: {row['support']:.4f} | Count: {row['count']:4d} | {{{pattern_str}}}")

    # Column frequency in patterns
    column_pattern_freq = {}
    for itemset in frequent_itemsets['itemsets']:
        for item in itemset:
            column_pattern_freq[item] = column_pattern_freq.get(item, 0) + 1

    top_columns = sorted(column_pattern_freq.items(), key=lambda x: x[1], reverse=True)[:20]
    print(f"\nTop 20 columns appearing in frequent patterns:")
    for col, freq in top_columns:
        print(f"  {col}: appears in {freq} patterns")

def main_full_column_analysis():
    """
    Main analysis preserving all 900 columns
    """
    print("FREQUENT PATTERN MINING - FULL 900 COLUMNS")
    print("=" * 60)

    # Create dataset
    print("Step 1: Creating 900-column dataset...")
    X_train = create_memory_efficient_dataset(n_rows=10000, n_cols=200, sparsity=0.95)

    # Parameter tuning
    print("\nStep 2: Parameter optimization...")
    min_support, max_len = adaptive_parameter_tuning(X_train)

    # Pattern mining on full dataset
    print("\nStep 3: Mining patterns across all 900 columns...")
    patterns = optimized_fpgrowth_full_columns(X_train, min_support=min_support, max_len=max_len)

    if patterns is not None:
        # Comprehensive analysis
        print("\nStep 4: Comprehensive analysis...")
        comprehensive_pattern_analysis(patterns, X_train)

        # Save results
        print("\nStep 5: Saving results...")
        patterns.to_csv('full_900_column_patterns.csv', index=False)
        print("Results saved to full_900_column_patterns.csv")

        return X_train, patterns
    else:
        print("Consider adjusting parameters or using batch processing")
        return X_train, None

# Alternative: Batch processing for very large pattern spaces
def batch_processing_approach(df, batch_size=200, min_support=0.005):
    """
    Process 900 columns in batches to manage memory
    """
    print(f"Processing 900 columns in batches of {batch_size}")

    all_patterns = []
    n_cols = len(df.columns)

    for i in range(0, n_cols, batch_size):
        end_idx = min(i + batch_size, n_cols)
        batch_cols = df.columns[i:end_idx]

        print(f"Processing batch {i//batch_size + 1}: columns {i+1}-{end_idx}")

        batch_df = df[batch_cols]
        batch_patterns = optimized_fpgrowth_full_columns(batch_df, min_support=min_support, max_len=4)

        if batch_patterns is not None:
            batch_patterns['batch'] = i//batch_size + 1
            all_patterns.append(batch_patterns)

    if all_patterns:
        combined_patterns = pd.concat(all_patterns, ignore_index=True)
        print(f"Total patterns found across all batches: {len(combined_patterns)}")
        return combined_patterns

    return None

if __name__ == "__main__":
    # Full analysis preserving all 900 columns
    X_train, patterns = main_full_column_analysis()

    # If memory issues occur, try batch processing
    if patterns is None:
        print("\nTrying batch processing approach...")
        batch_patterns = batch_processing_approach(X_train, batch_size=150)
```

```python
import pandas as pd
import numpy as np
from collections import defaultdict, Counter
import time
from typing import List, Dict, Tuple, Set
import warnings
warnings.filterwarnings('ignore')

# Try to import GPU libraries
try:
    import cupy as cp
    import cudf
    GPU_AVAILABLE = True
    print("GPU libraries available - will use GPU acceleration")
except ImportError:
    GPU_AVAILABLE = False
    print("GPU libraries not available - using CPU optimization")

# Try to import mlxtend for FP-Growth
try:
    from mlxtend.frequent_patterns import fpgrowth, association_rules
    from mlxtend.preprocessing import TransactionEncoder
    MLXTEND_AVAILABLE = True
except ImportError:
    MLXTEND_AVAILABLE = False
    print("mlxtend not available - using custom implementation")

class OptimizedPatternMiner:
    """
    Optimized frequent pattern mining for one-hot encoded data
    """

    def __init__(self, use_gpu=True, min_support=0.01, max_len=5):
        self.use_gpu = use_gpu and GPU_AVAILABLE
        self.min_support = min_support
        self.max_len = max_len
        self.patterns = {}

    def preprocess_data(self, X_train: pd.DataFrame) -> pd.DataFrame:
        """
        Preprocess the one-hot encoded data for pattern mining
        """
        print("Preprocessing data...")
        start_time = time.time()

        if self.use_gpu:
            # Convert to GPU dataframe for faster operations
            X_gpu = cudf.from_pandas(X_train)
            # Filter out columns with very low support
            col_sums = X_gpu.sum()
            min_count = int(self.min_support * len(X_gpu))
            valid_cols = col_sums[col_sums >= min_count].index.to_pandas()
            X_filtered = X_gpu[valid_cols].to_pandas()
        else:
            # CPU optimization
            col_sums = X_train.sum()
            min_count = int(self.min_support * len(X_train))
            valid_cols = col_sums[col_sums >= min_count].index
            X_filtered = X_train[valid_cols].copy()

        print(f"Filtered from {X_train.shape[1]} to {X_filtered.shape[1]} columns")
        print(f"Preprocessing completed in {time.time() - start_time:.2f} seconds")

        return X_filtered

    def convert_to_transactions(self, X: pd.DataFrame) -> List[List[str]]:
        """
        Convert one-hot encoded data to transaction format
        """
        print("Converting to transaction format...")
        start_time = time.time()

        transactions = []

        if self.use_gpu:
            # GPU-accelerated conversion
            X_gpu = cudf.from_pandas(X)
            for i in range(len(X_gpu)):
                row = X_gpu.iloc[i]
                transaction = row[row == 1].index.to_pandas().tolist()
                if transaction:  # Only add non-empty transactions
                    transactions.append(transaction)
        else:
            # Optimized CPU conversion using numpy
            X_array = X.values
            col_names = X.columns.tolist()

            for i in range(len(X_array)):
                indices = np.where(X_array[i] == 1)[0]
                if len(indices) > 0:
                    transaction = [col_names[j] for j in indices]
                    transactions.append(transaction)

        print(f"Converted to {len(transactions)} transactions in {time.time() - start_time:.2f} seconds")
        return transactions

    def find_frequent_patterns_mlxtend(self, X: pd.DataFrame) -> pd.DataFrame:
        """
        Use mlxtend's FP-Growth implementation (most efficient)
        """
        print("Finding frequent patterns using mlxtend FP-Growth...")
        start_time = time.time()

        # mlxtend expects boolean DataFrame
        X_bool = X.astype(bool)

        # Apply FP-Growth
        frequent_itemsets = fpgrowth(X_bool,
                                   min_support=self.min_support,
                                   use_colnames=True,
                                   max_len=self.max_len,
                                   verbose=1)

        # Sort by support
        frequent_itemsets = frequent_itemsets.sort_values('support', ascending=False)

        print(f"Found {len(frequent_itemsets)} frequent patterns in {time.time() - start_time:.2f} seconds")
        return frequent_itemsets

    def find_frequent_patterns_custom(self, transactions: List[List[str]]) -> Dict:
        """
        Custom implementation using optimized counting
        """
        print("Finding frequent patterns using custom implementation...")
        start_time = time.time()

        n_transactions = len(transactions)
        min_count = int(self.min_support * n_transactions)

        # Count single items first
        item_counts = Counter()
        for transaction in transactions:
            for item in transaction:
                item_counts[item] += 1

        # Filter frequent single items
        frequent_items = {item for item, count in item_counts.items() if count >= min_count}

        patterns = {}
        patterns[1] = {frozenset([item]): count/n_transactions
                      for item, count in item_counts.items() if item in frequent_items}

        # Generate patterns of increasing length
        for k in range(2, self.max_len + 1):
            patterns[k] = {}
            candidates = self.generate_candidates(list(patterns[k-1].keys()), k)

            # Count candidates
            candidate_counts = defaultdict(int)
            for transaction in transactions:
                transaction_set = set(transaction)
                for candidate in candidates:
                    if candidate.issubset(transaction_set):
                        candidate_counts[candidate] += 1

            # Filter frequent patterns
            for candidate, count in candidate_counts.items():
                if count >= min_count:
                    patterns[k][candidate] = count / n_transactions

            if not patterns[k]:  # No frequent patterns of this length
                break

            print(f"Found {len(patterns[k])} frequent {k}-itemsets")

        print(f"Pattern mining completed in {time.time() - start_time:.2f} seconds")
        return patterns

    def generate_candidates(self, frequent_itemsets: List[frozenset], k: int) -> List[frozenset]:
        """
        Generate candidate itemsets of length k
        """
        candidates = []
        n = len(frequent_itemsets)

        for i in range(n):
            for j in range(i + 1, n):
                # Join itemsets if they differ by only one item
                union = frequent_itemsets[i] | frequent_itemsets[j]
                if len(union) == k:
                    candidates.append(union)

        return list(set(candidates))

    def format_results(self, patterns_data) -> pd.DataFrame:
        """
        Format results into a readable DataFrame
        """
        if isinstance(patterns_data, pd.DataFrame):
            # mlxtend format
            results = patterns_data.copy()
            results['pattern'] = results['itemsets'].apply(lambda x: ', '.join(sorted(list(x))))
            results['length'] = results['itemsets'].apply(len)
            return results[['pattern', 'support', 'length']].sort_values('support', ascending=False)
        else:
            # Custom format
            all_patterns = []
            for length, patterns in patterns_data.items():
                for pattern, support in patterns.items():
                    all_patterns.append({
                        'pattern': ', '.join(sorted(list(pattern))),
                        'support': support,
                        'length': length
                    })

            return pd.DataFrame(all_patterns).sort_values('support', ascending=False)

    def mine_patterns(self, X_train: pd.DataFrame) -> pd.DataFrame:
        """
        Main method to mine frequent patterns
        """
        print(f"Starting pattern mining on dataset: {X_train.shape}")
        print(f"Using GPU: {self.use_gpu}")
        print(f"Minimum support: {self.min_support}")
        print(f"Maximum pattern length: {self.max_len}")

        total_start = time.time()

        # Preprocess data
        X_filtered = self.preprocess_data(X_train)

        # Choose algorithm based on availability
        if MLXTEND_AVAILABLE:
            patterns_data = self.find_frequent_patterns_mlxtend(X_filtered)
        else:
            transactions = self.convert_to_transactions(X_filtered)
            patterns_data = self.find_frequent_patterns_custom(transactions)

        # Format results
        results = self.format_results(patterns_data)

        print(f"Total mining completed in {time.time() - total_start:.2f} seconds")
        print(f"Found {len(results)} frequent patterns")

        return results

def create_dummy_data(n_rows=10000, n_cols=900, sparsity=0.95):
    """
    Create dummy one-hot encoded data for testing
    """
    print(f"Creating dummy dataset: {n_rows} rows, {n_cols} columns")

    # Create sparse random data
    np.random.seed(42)
    data = np.random.choice([0, 1], size=(n_rows, n_cols), p=[sparsity, 1-sparsity])

    # Add some correlated patterns to make it interesting
    # Pattern 1: flag_1, flag_2, flag_3 often occur together
    pattern1_indices = np.random.choice(n_rows, size=int(n_rows * 0.1), replace=False)
    data[pattern1_indices, [0, 1, 2]] = 1

    # Pattern 2: flag_5, flag_10, flag_15 often occur together
    pattern2_indices = np.random.choice(n_rows, size=int(n_rows * 0.08), replace=False)
    data[pattern2_indices, [4, 9, 14]] = 1

    # Pattern 3: flag_20, flag_25 often occur together
    pattern3_indices = np.random.choice(n_rows, size=int(n_rows * 0.15), replace=False)
    data[pattern3_indices, [19, 24]] = 1

    columns = [f'flag_{i+1}' for i in range(n_cols)]
    X_train = pd.DataFrame(data, columns=columns)

    print(f"Data created. Sparsity: {(data == 0).mean():.3f}")
    return X_train

def main():
    """
    Main execution function
    """
    print("=" * 60)
    print("OPTIMIZED FREQUENT PATTERN MINING")
    print("=" * 60)

    # Create dummy data
    X_train = create_dummy_data(n_rows=10000, n_cols=900, sparsity=0.95)

    # Initialize pattern miner
    miner = OptimizedPatternMiner(
        use_gpu=True,
        min_support=0.005,  # 0.5% minimum support
        max_len=4  # Maximum pattern length
    )

    # Mine patterns
    results = miner.mine_patterns(X_train)

    # Display results
    print("\n" + "=" * 60)
    print("TOP 20 MOST FREQUENT PATTERNS")
    print("=" * 60)
    print(results.head(20).to_string(index=False))

    # Pattern analysis
    print("\n" + "=" * 60)
    print("PATTERN ANALYSIS")
    print("=" * 60)
    print(f"Total patterns found: {len(results)}")
    print(f"Single item patterns: {len(results[results['length'] == 1])}")
    print(f"Two item patterns: {len(results[results['length'] == 2])}")
    print(f"Three+ item patterns: {len(results[results['length'] >= 3])}")

    print(f"\nHighest support pattern: {results.iloc[0]['pattern']}")
    print(f"Support: {results.iloc[0]['support']:.4f}")

    return results

if __name__ == "__main__":
    results = main()
```

```python
import pandas as pd
import numpy as np
from collections import defaultdict, Counter
import time
from typing import List, Dict, Tuple, Set
import warnings
warnings.filterwarnings('ignore')

# Try to import GPU libraries
try:
    import cupy as cp
    import cudf
    GPU_AVAILABLE = True
    print("GPU libraries available - will use GPU acceleration")
except ImportError:
    GPU_AVAILABLE = False
    print("GPU libraries not available - using CPU optimization")

# Try to import mlxtend for FP-Growth
try:
    from mlxtend.frequent_patterns import fpgrowth, association_rules
    from mlxtend.preprocessing import TransactionEncoder
    MLXTEND_AVAILABLE = True
except ImportError:
    MLXTEND_AVAILABLE = False
    print("mlxtend not available - using custom implementation")

class OptimizedPatternMiner:
    """
    Optimized frequent pattern mining for one-hot encoded data
    """

    def __init__(self, use_gpu=True, min_support=0.01, max_len=5):
        self.use_gpu = use_gpu and GPU_AVAILABLE
        self.min_support = min_support
        self.max_len = max_len
        self.patterns = {}

    def preprocess_data(self, X_train: pd.DataFrame, verbose: bool = True) -> pd.DataFrame:
        """
        Preprocess the one-hot encoded data for pattern mining
        """
        if verbose:
            print("Preprocessing data...")
        start_time = time.time()

        if self.use_gpu:
            # Convert to GPU dataframe for faster operations
            X_gpu = cudf.from_pandas(X_train)
            # Filter out columns with very low support
            col_sums = X_gpu.sum()
            min_count = int(self.min_support * len(X_gpu))
            valid_cols = col_sums[col_sums >= min_count].index.to_pandas()
            X_filtered = X_gpu[valid_cols].to_pandas()
        else:
            # CPU optimization
            col_sums = X_train.sum()
            min_count = int(self.min_support * len(X_train))
            valid_cols = col_sums[col_sums >= min_count].index
            X_filtered = X_train[valid_cols].copy()

        if verbose:
            print(f"Filtered from {X_train.shape[1]} to {X_filtered.shape[1]} columns")
            print(f"Preprocessing completed in {time.time() - start_time:.2f} seconds")

        return X_filtered

    def convert_to_transactions(self, X: pd.DataFrame) -> List[List[str]]:
        """
        Convert one-hot encoded data to transaction format
        """
        print("Converting to transaction format...")
        start_time = time.time()

        transactions = []

        if self.use_gpu:
            # GPU-accelerated conversion
            X_gpu = cudf.from_pandas(X)
            for i in range(len(X_gpu)):
                row = X_gpu.iloc[i]
                transaction = row[row == 1].index.to_pandas().tolist()
                if transaction:  # Only add non-empty transactions
                    transactions.append(transaction)
        else:
            # Optimized CPU conversion using numpy
            X_array = X.values
            col_names = X.columns.tolist()

            for i in range(len(X_array)):
                indices = np.where(X_array[i] == 1)[0]
                if len(indices) > 0:
                    transaction = [col_names[j] for j in indices]
                    transactions.append(transaction)

        print(f"Converted to {len(transactions)} transactions in {time.time() - start_time:.2f} seconds")
        return transactions

    def find_frequent_patterns_mlxtend(self, X: pd.DataFrame, verbose: bool = True) -> pd.DataFrame:
        """
        Use mlxtend's FP-Growth implementation (most efficient)
        """
        if verbose:
            print("Finding frequent patterns using mlxtend FP-Growth...")
        start_time = time.time()

        # mlxtend expects boolean DataFrame
        X_bool = X.astype(bool)

        # Apply FP-Growth
        frequent_itemsets = fpgrowth(X_bool,
                                   min_support=self.min_support,
                                   use_colnames=True,
                                   max_len=self.max_len,
                                   verbose=0)  # Set to 0 to suppress verbose output

        # Sort by support
        frequent_itemsets = frequent_itemsets.sort_values('support', ascending=False)

        if verbose:
            print(f"Found {len(frequent_itemsets)} frequent patterns in {time.time() - start_time:.2f} seconds")
        return frequent_itemsets

    def find_frequent_patterns_custom(self, transactions: List[List[str]]) -> Dict:
        """
        Custom implementation using optimized counting
        """
        print("Finding frequent patterns using custom implementation...")
        start_time = time.time()

        n_transactions = len(transactions)
        min_count = int(self.min_support * n_transactions)

        # Count single items first
        item_counts = Counter()
        for transaction in transactions:
            for item in transaction:
                item_counts[item] += 1

        # Filter frequent single items
        frequent_items = {item for item, count in item_counts.items() if count >= min_count}

        patterns = {}
        patterns[1] = {frozenset([item]): count/n_transactions
                      for item, count in item_counts.items() if item in frequent_items}

        # Generate patterns of increasing length
        for k in range(2, self.max_len + 1):
            patterns[k] = {}
            candidates = self.generate_candidates(list(patterns[k-1].keys()), k)

            # Count candidates
            candidate_counts = defaultdict(int)
            for transaction in transactions:
                transaction_set = set(transaction)
                for candidate in candidates:
                    if candidate.issubset(transaction_set):
                        candidate_counts[candidate] += 1

            # Filter frequent patterns
            for candidate, count in candidate_counts.items():
                if count >= min_count:
                    patterns[k][candidate] = count / n_transactions

            if not patterns[k]:  # No frequent patterns of this length
                break

            print(f"Found {len(patterns[k])} frequent {k}-itemsets")

        print(f"Pattern mining completed in {time.time() - start_time:.2f} seconds")
        return patterns

    def generate_candidates(self, frequent_itemsets: List[frozenset], k: int) -> List[frozenset]:
        """
        Generate candidate itemsets of length k
        """
        candidates = []
        n = len(frequent_itemsets)

        for i in range(n):
            for j in range(i + 1, n):
                # Join itemsets if they differ by only one item
                union = frequent_itemsets[i] | frequent_itemsets[j]
                if len(union) == k:
                    candidates.append(union)

        return list(set(candidates))

    def format_results(self, patterns_data) -> pd.DataFrame:
        """
        Format results into a readable DataFrame
        """
        if isinstance(patterns_data, pd.DataFrame):
            # mlxtend format
            results = patterns_data.copy()
            results['pattern'] = results['itemsets'].apply(lambda x: ', '.join(sorted(list(x))))
            results['length'] = results['itemsets'].apply(len)
            return results[['pattern', 'support', 'length']].sort_values('support', ascending=False)
        else:
            # Custom format
            all_patterns = []
            for length, patterns in patterns_data.items():
                for pattern, support in patterns.items():
                    all_patterns.append({
                        'pattern': ', '.join(sorted(list(pattern))),
                        'support': support,
                        'length': length
                    })

            return pd.DataFrame(all_patterns).sort_values('support', ascending=False)

    def mine_patterns(self, X_train: pd.DataFrame, verbose: bool = True) -> pd.DataFrame:
        """
        Main method to mine frequent patterns
        """
        if verbose:
            print(f"Starting pattern mining on dataset: {X_train.shape}")
            print(f"Using GPU: {self.use_gpu}")
            print(f"Minimum support: {self.min_support}")
            print(f"Maximum pattern length: {self.max_len}")

        total_start = time.time()

        # Preprocess data
        X_filtered = self.preprocess_data(X_train, verbose=verbose)

        # Choose algorithm based on availability
        if MLXTEND_AVAILABLE:
            patterns_data = self.find_frequent_patterns_mlxtend(X_filtered, verbose=verbose)
        else:
            transactions = self.convert_to_transactions(X_filtered)
            patterns_data = self.find_frequent_patterns_custom(transactions)

        # Format results
        results = self.format_results(patterns_data)

        if verbose:
            print(f"Total mining completed in {time.time() - total_start:.2f} seconds")
            print(f"Found {len(results)} frequent patterns")

        return results


    def analyze_results(self, results: pd.DataFrame, top_n: int = 10) -> None:
        """
        Analyze and display pattern mining results in a clean format
        """
        print("=" * 80)
        print("PATTERN MINING RESULTS ANALYSIS")
        print("=" * 80)

        # Basic statistics
        print(f"📊 SUMMARY STATISTICS")
        print(f"   Total patterns found: {len(results):,}")
        print(f"   Single flag patterns: {len(results[results['length'] == 1]):,}")
        print(f"   Multi-flag patterns: {len(results[results['length'] > 1]):,}")
        print(f"   Average support: {results['support'].mean():.4f}")
        print(f"   Highest support: {results['support'].max():.4f}")

        # Pattern length distribution
        print(f"\n📏 PATTERN LENGTH DISTRIBUTION")
        length_counts = results['length'].value_counts().sort_index()
        for length, count in length_counts.items():
            print(f"   {length}-flag patterns: {count:,}")

        # Top patterns by support
        print(f"\n🔥 TOP {top_n} MOST FREQUENT PATTERNS")
        print("-" * 80)
        for i, row in results.head(top_n).iterrows():
            support_pct = row['support'] * 100
            print(f"   {i+1:2d}. {row['pattern']:<50} | Support: {support_pct:5.1f}% | Length: {row['length']}")

        # Top multi-flag patterns
        multi_flag = results[results['length'] > 1]
        if len(multi_flag) > 0:
            print(f"\n🔗 TOP {min(top_n, len(multi_flag))} MULTI-FLAG PATTERNS")
            print("-" * 80)
            for i, row in multi_flag.head(top_n).iterrows():
                support_pct = row['support'] * 100
                print(f"   {i+1:2d}. {row['pattern']:<50} | Support: {support_pct:5.1f}% | Length: {row['length']}")

        print("=" * 80)


def create_dummy_data(n_rows=1000, n_cols=200, sparsity=0.95):
    """
    Create dummy one-hot encoded data for testing
    """
    print(f"Creating dummy dataset: {n_rows} rows, {n_cols} columns")

    # Create sparse random data
    np.random.seed(42)
    data = np.random.choice([0, 1], size=(n_rows, n_cols), p=[sparsity, 1-sparsity])

    # Add some correlated patterns to make it interesting
    # Pattern 1: flag_1, flag_2, flag_3 often occur together
    pattern1_indices = np.random.choice(n_rows, size=int(n_rows * 0.1), replace=False)
    data[pattern1_indices, 0] = 1  # flag_1
    data[pattern1_indices, 1] = 1  # flag_2
    data[pattern1_indices, 2] = 1  # flag_3

    # Pattern 2: flag_5, flag_10, flag_15 often occur together
    pattern2_indices = np.random.choice(n_rows, size=int(n_rows * 0.08), replace=False)
    data[pattern2_indices, 4] = 1   # flag_5
    data[pattern2_indices, 9] = 1   # flag_10
    data[pattern2_indices, 14] = 1  # flag_15

    # Pattern 3: flag_20, flag_25 often occur together
    pattern3_indices = np.random.choice(n_rows, size=int(n_rows * 0.15), replace=False)
    data[pattern3_indices, 19] = 1  # flag_20
    data[pattern3_indices, 24] = 1  # flag_25

    columns = [f'flag_{i+1}' for i in range(n_cols)]
    X_train = pd.DataFrame(data, columns=columns)

    print(f"Data created. Sparsity: {(data == 0).mean():.3f}")
    return X_train

def main():
    """
    Main execution function
    """
    print("=" * 60)
    print("OPTIMIZED FREQUENT PATTERN MINING")
    print("=" * 60)

    # Create dummy data
    X_train = create_dummy_data(n_rows=100, n_cols=900, sparsity=0.95)

    # Initialize pattern miner
    miner = OptimizedPatternMiner(
        use_gpu=False,
        min_support=0.005,  # 0.5% minimum support
        max_len=4  # Maximum pattern length
    )

    # Mine patterns with minimal output
    results = miner.mine_patterns(X_train, verbose=False)

    # Use the new analysis function for clean output
    miner.analyze_results(results, top_n=15)

    return results

if __name__ == "__main__":
    results = main()
```

```python
import shap

class ACeDeCShapAnalyzerGPU:
    """
    GPU-optimized SHAP analyzer for ACeDeC models with one-hot encoded data
    """

    def __init__(self, model, feature_names=None):
        self.model = model
        self.feature_names = feature_names
        self.shap_values = None
        self.explainer = None
        self.background_data = None
        self.sample_data = None

        # Detect and set model device
        if hasattr(model, 'neural_network') and model.neural_network is not None:
            try:
                first_param = next(model.neural_network.parameters())
                self.device = first_param.device
                print(f"✅ Detected model device: {self.device}")
            except:
                self.device = torch.device('cuda:0' if torch.cuda.is_available() else 'cpu')
                print(f"⚠️  Could not detect device, using: {self.device}")
        else:
            self.device = torch.device('cuda:0' if torch.cuda.is_available() else 'cpu')

    def reconstruction_error_function(self, x):
        """
        GPU-optimized reconstruction error function for SHAP analysis
        """
        # Convert to tensor if needed
        if not isinstance(x, torch.Tensor):
            x = torch.FloatTensor(x)

        # Move tensor to GPU
        x = x.to(self.device)

        self.model.neural_network.eval()
        with torch.no_grad():
            # Forward pass on GPU
            reconstructed = self.model.neural_network(x)
            mse_loss = F.mse_loss(reconstructed, x, reduction='none')
            # Move result back to CPU for SHAP processing
            return mse_loss.mean(dim=1).cpu().numpy()

    def prepare_data_gpu(self, X, background_size=50, sample_size=20):
        """
        Prepare data for GPU-optimized SHAP analysis
        """
        print(f"📊 Preparing data for GPU SHAP analysis...")

        # Ensure X is numpy array (CPU) for SHAP compatibility
        if isinstance(X, torch.Tensor):
            X = X.cpu().numpy()

        # Use appropriate sizes for GPU memory management
        background_size = min(background_size, len(X))
        sample_size = min(sample_size, len(X))

        # Random sampling
        background_indices = np.random.choice(len(X), background_size, replace=False)
        sample_indices = np.random.choice(len(X), sample_size, replace=False)

        self.background_data = X[background_indices].astype(np.float32)
        self.sample_data = X[sample_indices].astype(np.float32)

        print(f"   Background size: {len(self.background_data)}")
        print(f"   Sample size: {len(self.sample_data)}")
        print(f"   Data will be processed on: {self.device}")

        return self.background_data, self.sample_data

    def calculate_shap_values_gpu(self, X, background_size=50, sample_size=20):
        """
        Calculate SHAP values with GPU optimization
        """
        print("🔄 Calculating SHAP values (GPU-optimized)...")

        try:
            # Prepare data
            self.prepare_data_gpu(X, background_size, sample_size)

            # Create SHAP explainer
            print("   Creating GPU-compatible SHAP explainer...")
            self.explainer = shap.KernelExplainer(
                self.reconstruction_error_function,
                self.background_data
            )

            # Calculate SHAP values with progress monitoring
            print("   Computing SHAP values on GPU (this may take a while)...")
            print(f"   Processing {len(self.sample_data)} samples...")

            self.shap_values = self.explainer.shap_values(self.sample_data)

            print(f"✅ SHAP values calculated successfully on GPU!")
            print(f"   Shape: {np.array(self.shap_values).shape}")
            print(f"   Memory usage: {np.array(self.shap_values).nbytes / 1024**2:.2f} MB")

            return self.shap_values, self.explainer

        except Exception as e:
            print(f"❌ GPU SHAP calculation failed: {e}")
            print("💡 Try reducing background_size and sample_size for GPU memory")

            # Fallback with smaller sizes
            try:
                print("🔄 Trying with reduced sizes...")
                return self.calculate_shap_values_gpu(X, background_size=20, sample_size=10)
            except Exception as e2:
                print(f"❌ Fallback also failed: {e2}")
                return None, None

    def plot_shap_summary_gpu(self, plot_type='bar', max_display=20, figsize=(12, 8)):
        """
        Create GPU-optimized SHAP summary plot
        """
        if self.shap_values is None:
            print("❌ No SHAP values available")
            return

        try:
            plt.figure(figsize=figsize)

            if plot_type == 'bar':
                shap.summary_plot(
                    self.shap_values,
                    self.sample_data,
                    feature_names=self.feature_names,
                    plot_type="bar",
                    max_display=max_display,
                    show=False
                )
                plt.title("SHAP Feature Importance for Reconstruction Error (GPU Analysis)")

            elif plot_type == 'dot':
                shap.summary_plot(
                    self.shap_values,
                    self.sample_data,
                    feature_names=self.feature_names,
                    max_display=max_display,
                    show=False
                )
                plt.title("SHAP Summary: Feature Impact on Reconstruction Error (GPU Analysis)")

            plt.tight_layout()
            plt.show()

        except Exception as e:
            print(f"❌ GPU plotting failed: {e}")

    def get_top_problematic_features_gpu(self, top_n=20):
        """
        Get most problematic features from GPU-processed SHAP values
        """
        if self.shap_values is None:
            print("❌ No SHAP values available")
            return None

        # Calculate mean absolute SHAP values
        mean_shap = np.mean(np.abs(self.shap_values), axis=0)

        # Get top features
        top_indices = np.argsort(mean_shap)[-top_n:][::-1]

        if self.feature_names:
            top_features = [self.feature_names[i] for i in top_indices]
        else:
            top_features = [f'Feature_{i}' for i in top_indices]

        results = pd.DataFrame({
            'Feature': top_features,
            'SHAP_Importance': mean_shap[top_indices],
            'Feature_Index': top_indices
        })

        print(f"🔴 Top {top_n} Most Problematic Features (GPU Analysis):")
        print(results.to_string(index=False))

        return results

    def analyze_gpu_memory_usage(self):
        """
        Analyze GPU memory usage during SHAP analysis
        """
        if torch.cuda.is_available():
            print("📊 GPU Memory Analysis:")
            print(f"   Allocated: {torch.cuda.memory_allocated()/1024**2:.2f} MB")
            print(f"   Cached: {torch.cuda.memory_reserved()/1024**2:.2f} MB")
            print(f"   Max allocated: {torch.cuda.max_memory_allocated()/1024**2:.2f} MB")
        else:
            print("❌ CUDA not available for memory analysis")

# Initialize GPU-optimized analyzer
if model_is_valid:
    analyzer_gpu = ACeDeCShapAnalyzerGPU(model, feature_names)
    print("✅ GPU SHAP Analyzer initialized successfully!")
else:
    print("❌ Cannot initialize GPU SHAP analyzer - model training failed")
```

```python
def calculate_reconstruction_loss_gpu(model, X, batch_size=64):
    """
    Calculate reconstruction loss with GPU optimization and batch processing
    """
    print("🔄 Calculating reconstruction loss on GPU...")

    # Detect model device
    try:
        first_param = next(model.neural_network.parameters())
        device = first_param.device
        print(f"   Model device: {device}")
    except:
        device = torch.device('cuda:0' if torch.cuda.is_available() else 'cpu')
        print(f"   Using device: {device}")

    # Convert to tensor
    if not isinstance(X, torch.Tensor):
        X_tensor = torch.FloatTensor(X)
    else:
        X_tensor = X

    # Set to evaluation mode
    model.neural_network.eval()

    # Process in batches for GPU memory efficiency
    n_samples = len(X_tensor)
    all_losses = []
    all_reconstructed = []

    with torch.no_grad():
        for i in range(0, n_samples, batch_size):
            batch_end = min(i + batch_size, n_samples)
            batch_X = X_tensor[i:batch_end].to(device)

            # Forward pass
            batch_reconstructed = model.neural_network(batch_X)

            # Calculate MSE loss for batch
            batch_mse = F.mse_loss(batch_reconstructed, batch_X, reduction='none')

            # Store results (move back to CPU)
            all_losses.append(batch_mse.cpu())
            all_reconstructed.append(batch_reconstructed.cpu())

    # Concatenate all batches
    mse_loss = torch.cat(all_losses, dim=0)
    reconstructed = torch.cat(all_reconstructed, dim=0)

    # Calculate statistics
    total_loss = mse_loss.mean().item()
    sample_errors = mse_loss.mean(dim=1).numpy()
    feature_errors = mse_loss.mean(dim=0).numpy()

    print(f"✅ GPU reconstruction loss calculated successfully!")
    print(f"   Total Loss: {total_loss:.6f}")
    print(f"   Worst Sample Error: {sample_errors.max():.6f}")
    print(f"   Best Sample Error: {sample_errors.min():.6f}")
    print(f"   Most Problematic Feature Error: {feature_errors.max():.6f}")

    # GPU memory cleanup
    if torch.cuda.is_available():
        torch.cuda.empty_cache()
        print(f"   GPU memory cleared")

    return {
        'total_loss': total_loss,
        'sample_errors': sample_errors,
        'feature_errors': feature_errors,
        'reconstructed': reconstructed.numpy(),
        'mse_matrix': mse_loss.numpy()
    }

# Calculate reconstruction loss with GPU optimization
reconstruction_results = calculate_reconstruction_loss_gpu(model, X, batch_size=64)
```

```python
def run_complete_gpu_analysis(model, X, feature_names, analyzer_gpu):
    """
    Run complete analysis pipeline optimized for GPU processing
    """
    print("🚀 Running Complete ACeDeC + SHAP Analysis (GPU-Optimized)")
    print("=" * 70)

    # GPU Memory monitoring
    if torch.cuda.is_available():
        print(f"Initial GPU Memory: {torch.cuda.memory_allocated()/1024**2:.2f} MB")

    # Step 1: Calculate reconstruction loss on GPU
    print("\n1. CALCULATING RECONSTRUCTION LOSS ON GPU")
    try:
        reconstruction_results = calculate_reconstruction_loss_gpu(model, X, batch_size=64)

        # Find most problematic features by reconstruction error
        feature_losses = reconstruction_results['feature_errors']
        worst_features_idx = np.argsort(feature_losses)[-10:][::-1]
        print(f"\n🔴 Top 10 features with highest reconstruction error:")
        for i, idx in enumerate(worst_features_idx):
            print(f"   {i+1}. {feature_names[idx]}: {feature_losses[idx]:.6f}")

    except Exception as e:
        print(f"❌ GPU reconstruction loss calculation failed: {e}")
        return False

    # Step 2: GPU-optimized SHAP Analysis
    print("\n2. GPU-OPTIMIZED SHAP ANALYSIS")
    try:
        # Use optimized sample sizes for GPU
        shap_values, explainer = analyzer_gpu.calculate_shap_values_gpu(
            X, background_size=30, sample_size=10
        )

        if shap_values is not None:
            # Analyze GPU memory usage
            analyzer_gpu.analyze_gpu_memory_usage()

            # Get top problematic features
            top_features = analyzer_gpu.get_top_problematic_features_gpu(top_n=15)

            # Create GPU-optimized visualizations
            print("\n3. CREATING GPU-OPTIMIZED VISUALIZATIONS")
            analyzer_gpu.plot_shap_summary_gpu(plot_type='bar', max_display=15)

            # GPU memory cleanup
            if torch.cuda.is_available():
                torch.cuda.empty_cache()
                print("✅ GPU memory cleaned up")

            return True
        else:
            print("❌ GPU SHAP analysis failed")
            return False

    except Exception as e:
        print(f"❌ GPU SHAP analysis failed: {e}")
        print(f"   Error details: {str(e)}")
        return False

# Run the complete GPU-optimized analysis
if model_is_valid and 'analyzer_gpu' in locals():
    success = run_complete_gpu_analysis(model, X, feature_names, analyzer_gpu)
    if success:
        print("\n🎉 GPU-optimized analysis completed successfully!")
    else:
        print("\n⚠️  GPU analysis completed with some errors")
else:
    print("❌ Cannot run GPU analysis - prerequisites not met")
```
